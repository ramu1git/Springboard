Checking the number of missing values (char '?' in the data) in each feature

```python
for i in range(1,65):
    missing = len(poli[poli['X'+str(i)]=='?'])
    if missing > 0:
        print('X'+str(i),':',missing)
```

```python
poli.replace(to_replace='?',value=np.nan,inplace=True)
```

```python
poli = poli.drop('X37',axis=1)
```

```python
poli = poli.astype(float)
```

Replacing the missing values in all features with median of corresponding feature

```python
#poli.fillna(0,inplace=True)
for i in range(1,65):
    if i != 37:
        poli['X'+str(i)] = poli['X'+str(i)].fillna(poli['X'+str(i)].median())
```

Z-normalization

```python
for i in range (1,65):
    if i != 37:
        for j in range(0,10503):
            poli['X'+str(i)][j] = (poli['X'+str(i)][j]-poli['X'+str(i)].mean())/poli['X'+str(i)].std()
```

There are many outliers in each feature. As the number of features are more (63), outliers are being identified observation wise (instead of each feature wise) using Euclidean distance between each observation and the Centroid.

```python
centroid = []
for i in range(1,65):
    if i != 37:
        col = poli['X'+str(i)]
        centroid.append(col.mean())
print('centroid:', centroid)
```

```python
polix = poli.drop('y',axis=1)
```

```python
poliy = poli['y']
```

```python
from scipy.spatial import distance

eucli_dist = []

for i in range(0,10503):
    row = polix[i:i+1]
    dist = distance.euclidean(row, centroid)
    eucli_dist.append(dist)
```

Boxplot of the euclidean distance between each observation and centroid

```python
sns.boxplot(eucli_dist)
plt.show()
```

Identifying the outliers beyond 1.5*IQR

```python
def get_iqr_cutoffs1(feature,upper_cutoff=1.5,lower_cutoff=1.5):
    firstq = feature.quantile(0.25)
    thirdq = feature.quantile(0.75)
    iqr = thirdq - firstq
    lower_bound1 = firstq - lower_cutoff * iqr
    upper_bound1 = thirdq + upper_cutoff * iqr
    
    return lower_bound1, upper_bound1
```

```python
get_iqr_cutoffs1(pd.DataFrame(eucli_dist))
```

```python
eucli_df = pd.DataFrame(eucli_dist)
```

```python
eucli_df[eucli_df[0]>4.99]
```

Identifying the outliers beyond 3*IQR

```python
def get_iqr_cutoffs2(feature,upper_cutoff=3,lower_cutoff=3):
    firstq = feature.quantile(0.25)
    thirdq = feature.quantile(0.75)
    iqr = thirdq - firstq
    lower_bound2 = firstq - lower_cutoff * iqr
    upper_bound2 = thirdq + upper_cutoff * iqr
    
    return lower_bound2, upper_bound2
```

```python
get_iqr_cutoffs2(pd.DataFrame(eucli_dist))
```

```python
eucli_df[eucli_df[0]>7.02]
```
